```python
import pandas as pd

esg_only_df = pd.read_csv('ESG_only_ECT_edited.csv')
esg_only_df.head()
```

```python
esg_list = esg_only_df.Source_File.unique()
esg_list.shape
```

```python
esg_list[0]
```

```python
sample = pd.read_csv('df/'+esg_list[0])
sample = sample.dropna()
sample[['Company', 'Position']] = sample['Position'].str.split(' - ', n=1, expand=True)
sample = sample[['Company','Position','Speaker']]
sample = sample.drop_duplicates()
sample['file_name'] = esg_list[0]
sample.head()
```

```python
cleaned_participant = []

for i in esg_list:
    sample = pd.read_csv('df/'+i)
    sample = sample.dropna()
   
    if (len(sample)!=0):
        split_data = sample['Position'].str.split(' - ', expand=True, n=1).apply(lambda x: x.str.strip() if x is not None else x)

        if (len(split_data.columns)==1):
            split_data.loc[split_data[0] == "", 0] = None
            sample['Company'] = split_data
            sample['Co-Position'] = None
        else:
            split_data.loc[split_data[0] == "", 0] = None
            split_data.loc[split_data[1] == "", 0] = None
            sample[['Company', 'Co-Position']] = split_data

        sample = sample[['Company','Co-Position','Speaker']]
        sample = sample.drop_duplicates()
        sample['file_name'] = i
        cleaned_participant.append(sample)
    
    print(i)
    
cleaned_participant_df = pd.concat(cleaned_participant, ignore_index=True)
```
